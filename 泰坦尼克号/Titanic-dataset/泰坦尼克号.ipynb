```python
import pandas as pd 
import numpy as np
from pandas import  Series,DataFrame
```

```python
train = pd.read_csv('train.csv') # 训练集
test = pd.read_csv('test.csv') # 测试集

```

```python
all = train.append(test,ignore_index = True)  # 合并数据集，方便一起清洗
all.info # 观测数据集的基本信息
all.head
```

```python
# 进行数据清洗
# 填充年龄和船票价格 ,如果缺失值是数据类型，那么就使用平均数，如果是字符串类型，那么就是要出现最多的字符串代替
all.Age = all.Age.fillna(all.Age.mean())
all.Fare = all.Fare.fillna(all.Fare.mean())

# 填充登录陆港口，最多的港口为s
all.Embarked = all.Embarked.fillna('S')

#因为座位值缺失过多，那么不可以遵守这样的规则，所以只可以填充U 来作为不知道
all.Cabin = all.Cabin.fillna('U')
```

```python
all
```

```python
# 进行one-hot编码
sexdict ={"male":1,"female":0}
sex = all.Sex.map(sexdict)

# 改变港口的编码，应用get_demmies,get_demmies将类名变量向量化
embarked = pd.DataFrame()
embarked = pd.get_dummies(all.Embarked,prefix= 'Embarked') #prefix是加上前缀

#对船舱登记用同样的方法编码
pclass = pd.DataFrame()
pclass = pd.get_dummies(all.Pclass,prefix = 'Pclass')
```

```python
all.head()
```

```python
# 对姓名进行分析，提取姓名中的头衔特征
# 观察数据后发现前面是逗号，后面是点号，
#建立函数进行分割
def title1(name):
    name1 = name.split(',')[1]
    name2 = name1.split('.')[0]
    #strip函数移除字符串头尾指定的字符
    name3 = name2.strip()
    return name3

#应用函数
titiee = pd.DataFrame()
titiee['Title'] = all.Name.map(title1)

#因为头衔太多了，所以要分类映射一下
#这里需要统计一共出现过多少个头衔
#然后要知道这些头衔在国外代表什么意思
#最后将不同的头衔映射为一样的
titleDict = {
                    "Capt":       "Officer",
                    "Col":        "Officer",
                    "Major":      "Officer",
                    "Jonkheer":   "Royalty",
                    "Don":        "Royalty",
                    "Sir" :       "Royalty",
                    "Dr":         "Officer",
                    "Rev":        "Officer",
                    "the Countess":"Royalty",
                    "Dona":       "Royalty",
                    "Mme":        "Mrs",
                    "Mlle":       "Miss",
                    "Ms":         "Mrs",
                    "Mr" :        "Mr",
                    "Mrs" :       "Mrs",
                    "Miss" :      "Miss",
                    "Master" :    "Master",
                    "Lady" :      "Royalty"
                    }
#把每一个头衔都映射一遍
titiee['Title'] = titiee['Title'].map(titleDict)

#再把头衔变成onehot编码
titiee = pd.get_dummies(titiee['Title'])

titiee.head()

```

```python
# 接下来从座位号中提取坐席类别，坐席类别就是座位号的首字母，假如为U的话则代表不知道。 对船上每个乘客的家庭成员数量进行一个统计。 最后将原始数据复制一份，该修改的修改，该删除的列删除，变成一个清洗好的数据。
seat = pd.DataFrame()
seat['seat'] = all['Cabin']
# 提前座位号的首字母特征
seat.seat = seat.seat.map(lambda a :a[0])
# onehot 编码
seat = pd.get_dummies(seat.seat,prefix='Cabin')  # 将数据和原始数据进行合并 
seat.head()

```

```python
family = pd.DataFrame()
family['family'] = all.SibSp + all.Parch + 1
family.head()
```

```python
# 将数据复制一份 
da = train.append(test,ignore_index = True)
#制作同样的修改
da.Age = da.Age.fillna(da.Age.mean())
da.Fare = da.Fare.fillna(da.Fare.mean())
da.Embarked = da.Embarked.fillna('S')
da.Cabin = da.Cabin.fillna('U')
#删除不需要的列
da = da.drop(['Name','Pclass','Cabin', 'Sex', 'Parch', 'Embarked', 'SibSp', 'Ticket'], axis =1)
#合并修改后的列
da = pd.concat([da,sex,pclass,embarked,titiee,seat,family],axis=1)
da.head()

```

```python
# 建模
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(all.corr())
plt.show()
```

```python
# 切分训练集
#把训练数据集切分出来
source = da.loc[0:890, :]
pred = da.loc[890:, :]
source = source.drop(['PassengerId'], axis =1 )
```

```python
# 导入逻辑回归算法
from sklearn.linear_model import  LogisticRegression
#导入交叉验证的包
from sklearn.cross_validation import  KFold
# 假设特征为年龄和船票
predictors = ['Age', 'Fare']

#建立模型
model = LogisticRegression()
#做交叉验证
kf = KFold(source.shape[0], n_folds = 3, random_state = 1)

for train, test in kf:
    train_predictors = (source[predictors].iloc[train, :])
    train_target = source['Survived'].iloc[train]
    model.fit(train_predictors, train_target)

#测试一下这个模型的效果
model.score(source[predictors].iloc[test, :],source['Survived'].iloc[test] )
```
